```python
#EJERCICIO 1

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Definir la funcin exponencial
def exp_func(x, a, b):
    return b * np.exp(a * x)

# Definir la funcin de potencia
def power_func(x, a, b):
    return b * x**a

# Datos
xi = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
yi = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

# Polinomio de grado 1
coef_1 = np.polyfit(xi, yi, 1)
poly_1 = np.poly1d(coef_1)
error_1 = np.sum((poly_1(xi) - yi)**2)

plt.scatter(xi, yi, color='blue', label='Datos')
plt.plot(xi, poly_1(xi), color='red', label='Polinomio grado 1')
plt.xlabel('xi')
plt.ylabel('yi')
plt.title('Ajuste por mnimos cuadrados (grado 1)')
plt.legend()
plt.show()

print("Polinomio grado 1")
print("Coeficientes:", coef_1)
print("Error:", error_1)

# Polinomio de grado 2
coef_2 = np.polyfit(xi, yi, 2)
poly_2 = np.poly1d(coef_2)
error_2 = np.sum((poly_2(xi) - yi)**2)

plt.scatter(xi, yi, color='blue', label='Datos')
plt.plot(xi, poly_2(xi), color='red', label='Polinomio grado 2')
plt.xlabel('xi')
plt.ylabel('yi')
plt.title('Ajuste por mnimos cuadrados (grado 2)')
plt.legend()
plt.show()

print("Polinomio grado 2")
print("Coeficientes:", coef_2)
print("Error:", error_2)

# Polinomio de grado 3
coef_3 = np.polyfit(xi, yi, 3)
poly_3 = np.poly1d(coef_3)
error_3 = np.sum((poly_3(xi) - yi)**2)

plt.scatter(xi, yi, color='blue', label='Datos')
plt.plot(xi, poly_3(xi), color='red', label='Polinomio grado 3')
plt.xlabel('xi')
plt.ylabel('yi')
plt.title('Ajuste por mnimos cuadrados (grado 3)')
plt.legend()
plt.show()

print("Polinomio grado 3")
print("Coeficientes:", coef_3)
print("Error:", error_3)

# Funcin de la forma b e^(ax)
popt_exp, pcov_exp = curve_fit(exp_func, xi, yi)
error_exp = np.sum((exp_func(xi, *popt_exp) - yi)**2)

plt.scatter(xi, yi, color='blue', label='Datos')
plt.plot(xi, exp_func(xi, *popt_exp), color='red', label='Funcin exponencial')
plt.xlabel('xi')
plt.ylabel('yi')
plt.title('Ajuste por mnimos cuadrados (exponencial)')
plt.legend()
plt.show()

print("Funcin exponencial")
print("Parmetros:", popt_exp)
print("Error:", error_exp)

# Funcin de la forma b x^a
popt_power, pcov_power = curve_fit(power_func, xi, yi)
error_power = np.sum((power_func(xi, *popt_power) - yi)**2)

plt.scatter(xi, yi, color='blue', label='Datos')
plt.plot(xi, power_func(xi, *popt_power), color='red', label='Funcin de potencia')
plt.xlabel('xi')
plt.ylabel('yi')
plt.title('Ajuste por mnimos cuadrados (potencia)')
plt.legend()
plt.show()

print("Funcin de potencia")
print("Parmetros:", popt_power)
print("Error:", error_power)
```

```python
#EJERCICIO 3

import numpy as np
import matplotlib.pyplot as plt

# Datos del Ejercicio 3
act_scores = np.array([28, 29, 25, 28, 28, 27, 27, 29, 28, 21, 33, 28, 28, 28, 29, 26, 23, 30, 27, 24])
gpa_scores = np.array([3.84, 3.75, 3.21, 3.65, 3.23, 3.87, 3.63, 3.75, 3.75, 1.66, 3.20, 3.12, 3.41, 2.96, 3.38, 2.92, 3.53, 3.10, 2.03, 2.81])

# Ajuste lineal
coef_act = np.polyfit(act_scores, gpa_scores, 1)
poly_act = np.poly1d(coef_act)

# Calcular el error
error_act = np.sum((poly_act(act_scores) - gpa_scores)**2)

# Graficar
plt.scatter(act_scores, gpa_scores, color='blue', label='Datos')
plt.plot(act_scores, poly_act(act_scores), color='red', label='Recta de ajuste')
plt.xlabel('Puntuacin ACT')
plt.ylabel('Promedio de puntos')
plt.title('Ajuste por mnimos cuadrados')
plt.legend()
plt.show()

# Resultados del ajuste lineal
print("Coeficientes de la recta de ajuste (m, b):", coef_act)
print("Error de ajuste:", error_act)
```

```python
#EJERCICIO 4

import numpy as np
import matplotlib.pyplot as plt

# Datos del Ejercicio 4
weights = np.array([4800, 3700, 3400, 2800, 1900])
percentages = np.array([3.1, 4.0, 5.2, 6.4, 9.6])

# Ajuste lineal
coef_weight = np.polyfit(weights, percentages, 1)
poly_weight = np.poly1d(coef_weight)

# Calcular el error
error_weight = np.sum((poly_weight(weights) - percentages)**2)

# Graficar
plt.scatter(weights, percentages, color='blue', label='Datos')
plt.plot(weights, poly_weight(weights), color='red', label='Recta de ajuste')
plt.xlabel('Peso promedio (lb)')
plt.ylabel('Porcentaje de presentacin (%)')
plt.title('Ajuste por mnimos cuadrados')
plt.legend()
plt.show()

# Resultados del ajuste lineal
print("Coeficientes de la recta de ajuste (m, b):", coef_weight)
print("Error de ajuste:", error_weight)
```
